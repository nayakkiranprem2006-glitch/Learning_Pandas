````python
import pandas as pd
#1.  Sort this DataFrame by column 'Age' in ascending order:
#    python df = pd.DataFrame({"Name": ["Alice", "Bob", "Charlie"], "Age": [25, 30, 20]})

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
                   "Age": [25, 30, 20]
})

sort = df.sort_values("Age", ascending= True)
print(sort)
````

````python
# ➤ Q2: Sort the DataFrame by 'Name' in descending order.  
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
                   "Age": [25, 30, 20]
})

sort = df.sort_values("Name", ascending = False)
print(sort)
````

````python
# ➤ Q3:Select rows where Age > 20 and Name starts with 'A'.
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie","Sahil", "Anurag", "Pawan", "Aditi"],
     "Age": [25, 30, 20,48,65,26,19]
})

sort =df[(df["Age"]> 20) & (df["Name"].str.startswith("A"))]
print(sort)
````

````python
# ➤ Q4:Reset the index of the above DataFrame.
import numpy as np
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie","Sahil", "Anurag", "Pawan", "Aditi"],
     "Age": [25, 30, 20,48,65,26,19]
})

df.reset_index(drop = True, inplace = True)
print(df)
````

````python
# ➤ Q5:Set 'Name' column as the index.
df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie","Sahil", "Anurag", "Pawan", "Aditi"],
     "Age": [25, 30, 20,48,65,26,19],
    "Salary" : [10000,20000,30000,40000,50000,60000,70000]
})

df.set_index("Name", inplace = True)
print(df)
````

````python
# ➤ Q6:Group the DataFrame by 'Age' and find the count of each group.

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie","Sahil", "Anurag", "Pawan", "Aditi"],
     "Age": [25, 30, 20,25, 30,26,19],
    "Salary" : [10000,20000,30000,40000,50000,60000,70000]
})

group = df.groupby("Age").count()
print(group)
group = df.groupby("Age")["Name"].count()
print(group)
````

````python
# ➤ Q7:Use apply() to add 10 years to each age.

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie","Sahil", "Anurag", "Pawan", "Aditi"],
     "Age": [25, 30, 20,25, 30,26,19],
    "Salary" : [10000,20000,30000,40000,50000,60000,70000]
})

df["Age"] += 10
df
````

````python
# ➤ Q8:Replace 'Bob' with 'Robert' in the 'Name' column.

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie","Sahil", "Anurag", "Pawan", "Aditi"],
     "Age": [25, 30, 20,25, 30,26,19],
    "Salary" : [10000,20000,30000,40000,50000,60000,70000]
}) 

df.loc[1, "Name"] = "Robert"
df
````

````python
# 9. Concatenate two DataFrames vertically.

df1 = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
     "Age": [25, 30, 20],
    "Salary" : [10000,20000,30000]
}) 

df2 = pd.DataFrame({
    "Name": ["Sahil", "Anurag", "Pawan", "Aditi"],
     "Age": [25, 30,26,19],
    "Salary" : [40000,50000,60000,70000]
}) 

concatination = pd.concat([df1,df2], axis =0, ignore_index = True )
print(concatination)
````

````python
# 10. Create a pivot table from this DataFrame:  
#     ```python
data = {"Name": ["Alice", "Bob", "Alice", "Bob"],
        "Subject": ["Math", "Math", "Science", "Geography"],
        "Score": [85, 90, 95, 80]}
df = pd.DataFrame(data)

table1 = pd.pivot_table(df, index = ["Name", "Subject", "Score"])
print(f"table1 Format:\n{table1}\n")

table2 = pd.pivot_table(df,index = "Name", columns = "Subject", values = "Score")
print(f"\ntable2 Format:\n{table2}")
````
