```python
# ➤ Q1: Load a CSV file named data.csv using pandas.

import pandas as pd

df = pd.read_csv("Train.csv")

print(df)
```

```python
# Set Index
# ➤ Q2: Set the 'ID' column as index from a given DataFrame.

import numpy as np

df = pd.DataFrame(np.random.randint(1, 101, (10, 9)))  # 9 columns of data
df["ID"] = np.arange(1,11)   # Add "ID" column
print(df)
```

```python
# ➤ Q3: Select columns 'Name' and 'Age' from a DataFrame.
df = pd.DataFrame({
    "Name" : ["Atul", "Ramesh", "Soni", "Ajit"],
    "Age" : [54, 26, 42, 22],
    "City" : ["Delhi", "Bangluru", "Gujarat", "Ahamdabad"]
})
df[["Name", "Age"]]
```

```python
# ➤ Q4: Select rows where 'Age' is greater than 30.

df[df["Age"]>30]
```

```python
# ➤ Q5: Add a new column 'Score' with values.
df["Score"] = np.random.randint(30,95,len(df))
df
```

```python
# ➤ Q6: Update all 'Score' values to 100.
df["Score"] = 100
df
```

```python
# ➤ Q7: Drop the 'Gender' column from the DataFrame.
new_df = pd.DataFrame({
    "Name" : ["Atul", "Ramesh", "Soni", "Ajit"],
    "Age" : [54, 26, 42, 22],
    "City" : ["Delhi", "Bangluru", "Gujarat", "Ahamdabad"],
    "Gender" : ["Male", "Male" , "Female", "Male"]
})
new_df.drop("Gender", axis = 1, inplace= True)
new_df
```

```python

# ➤ Q8: Drop the row with index 2.

new_df.drop(2, axis = 0, inplace = True)
new_df
```

```python
# ➤ Q9: Check for missing values in each column.
new_df.iloc[1,1] = np.nan
new_df.isnull()
```

```python
# ➤ Q10: Fill missing values in 'Age' column with 0.
new_df["Age"].fillna(method = 'ffill', inplace = True)
new_df["Age"].fillna(0, inplace= True)
new_df
```
