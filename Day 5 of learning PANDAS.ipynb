```python
# ðŸ”¹Q1: Convert the 'Date' column to datetime and extract the month.  
# df = pd.DataFrame({'Date': ['2023-01-15', '2023-03-20', '2023-07-25']})

import pandas as pd
df = pd.DataFrame({'Date': ['2023-01-15', '2023-03-20', '2023-07-25']})
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month

print(df)
```

```python
# âž¤ Q2: Filter rows where the date is after '2023-03-01'.

df = pd.DataFrame({'Date': ['2023-01-15', '2023-03-20', '2023-07-25']})
df["Date"] = pd.to_datetime(df["Date"])

filtered = df[df["Date"]>pd.to_datetime("2023-03-01")]
print(filtered)
```

```python
# ðŸ”¹Q3: Group the DataFrame by 'Department' and calculate both mean and max of 'Salary'.

df = pd.DataFrame({
    "Name" : ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Department" : ["HR", "IT", "HR", "Finance", "IT"],
    "Salary" : [50000, 60000, 55000, 70000, 62000]
})

# Type 1 to solve
mean_data = df.groupby("Department")["Salary"].mean()
max_data = df.groupby("Department")["Salary"].max()
print("mean_data: ", mean_data)
print("max_data: ",max_data)
```

```python
# ðŸ”¹Q4: Group by 'Team' and apply agg() to compute both count and sum of 'Points'.

df = pd.DataFrame({
    "Name" : ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Team" : ["A", "B", "B", "A", "C"],
    "Points" : [30,80,60,30,90]
})

group = df.groupby("Team")["Points"].agg(["count", "sum"])
print(group)
```

```python
# ðŸ”¹Q5: Use apply() with a lambda to create a new column 'Age Group' (e.g., "Senior" if Age > 40 else "Junior").

df = pd.DataFrame({
    "Name" : ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age" : [45, 28, 39, 50, 33]
})

df["Age Group"] = df["Age"].apply(lambda age: "Senior" if age > 40 else"Junior")
print(df)
```

```python
# ðŸ”¹Q6: Filter rows where the 'Name' column starts with 'A'.

df = pd.DataFrame({
    "Name" : ["Alice", "Bob", "Charlie", "David", "Eva"]
})
Filter = df[df["Name"].str.startswith("A")]
print(Filter)
```

```python
# ðŸ”¹Q7: Remove duplicate rows based on the 'Email' column.

df = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", "Bob"],
    "Email": ["alice@example.com", "bob@example.com", "charlie@example.com", "david@example.com", "bob@example.com"]
})

df.drop_duplicates(subset = "Email",inplace = True)
df
```

```python
# ðŸ”¹Q8: Save a DataFrame to a CSV file named "employees.csv" without the index.

df = pd.DataFrame({
    "Employee_Name" : ["Alice", "Bob", "Charlie", "David"],
    "Department" : ["HR", "IT", "HR", "Finance"],
    "Salary" : [50000, 60000, 55000, 70000]
})

df.to_csv("employees.csv", index = False)
    
```

```python
# ðŸ”¹Q9: Calculate a rolling average of 'Sales' over the past 3 rows.

df = pd.DataFrame({
    "Date": pd.date_range(start="2023-01-01", periods=7, freq='D'),
    "Sales": [200, 220, 250, 270, 300, 320, 350]
})

df['Rolling_Avg'] = df['Sales'].rolling(window=3).mean()
print(df)
```

```python
# ðŸ”¹Q10: Count missing values in each column of the DataFrame.
import numpy as np

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', np.nan, 'David', 'Eva'],
    'Age': [25, np.nan, 30, 22, np.nan],
    'Salary': [50000, 60000, np.nan, 70000, 65000]
})

missing_counts = df.isnull().sum()
print(missing_counts)
```
